```python
from pathlib import Path

raw_dir = Path(r"C:\Users\athar\AI Projects\Li-Ion-Battery-Reliability-Degradation-Analytics\data\raw")

mat_files = sorted(raw_dir.rglob("*.mat"))

print("Total MAT files:", len(mat_files))

mat_cells = [f.stem for f in mat_files]

print("\nSample MAT cells:")
print(mat_cells[:10])
```

```python
interim_dir = Path(r"C:\Users\athar\AI Projects\Li-Ion-Battery-Reliability-Degradation-Analytics\data\interim")

discharge_parquets = sorted(interim_dir.glob("*_discharge.parquet"))
impedance_parquets = sorted(interim_dir.glob("*_impedance.parquet"))

print("Discharge parquet files:", len(discharge_parquets))
print("Impedance parquet files:", len(impedance_parquets))
```

```python
mat_cells = set(f.stem for f in mat_files)

discharge_cells = set(
    f.name.replace("_discharge.parquet", "")
    for f in discharge_parquets
)

missing_in_parquet = mat_cells - discharge_cells
extra_in_parquet = discharge_cells - mat_cells

print("MAT cells:", len(mat_cells))
print("Parquet cells:", len(discharge_cells))

print("\nMissing parquet conversions:")
print(missing_in_parquet)

print("\nUnexpected parquet files:")
print(extra_in_parquet)
```
